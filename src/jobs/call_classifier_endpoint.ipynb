```python
"""Call the classifier Model Serving endpoint (see resources/deploy_classifier_endpoint.yml).

Run after training + deploy jobs. Same feature schema as create_dummy_data.ipynb (33 numeric features).
"""

from __future__ import annotations

from databricks.sdk import WorkspaceClient

# Default matches databricks.yml `classifier_serving_endpoint_name`.
dbutils.widgets.text("endpoint_name", "flip_flopper_classifier_endpoint")

endpoint_name = dbutils.widgets.get("endpoint_name").strip()
if not endpoint_name:
    raise ValueError("endpoint_name must be non-empty")

N_FEATURES = 33
example_record = {f"feature_{i:03d}": 0.0 for i in range(N_FEATURES)}

w = WorkspaceClient()
response = w.serving_endpoints.query(
    name=endpoint_name,
    dataframe_records=[example_record],
)

print("endpoint:", endpoint_name)
print("request (one row):", example_record)
print("response:", response)
if getattr(response, "predictions", None) is not None:
    print("predictions:", response.predictions)
```
